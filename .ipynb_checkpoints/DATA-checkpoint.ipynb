```python
data_.coloumb["key"] = data_.coloumb["KEY"].apply(lambda x: x.split(".")[0])
data_.coloumb["m1"] = data_.coloumb["key"].apply(lambda x: x.split("_")[1])
data_.coloumb["m2"] = data_.coloumb["key"].apply(lambda x: x.split("_")[2])
data_.coloumb["key"] = data_.coloumb["key"].apply(lambda x: x.split("_")[0])
data_.coloumb

from ff_energy.cli import get_structures
from ff_energy.potential import Ecoloumb
structures, pdbs = get_structures("water_cluster")
structure_key_pairs = {p.split(".")[0]: s for p,s in zip(pdbs, structures)}

E_col_dcms = []
dist_dcms = []
min_hbond = []
angles_dcms = []
azi_dcms = []
dih_dcms = []
dcms_ = []


# def kabsch(mapping_points,true_points):
#     h = mapping_points.T @ true_points
#     u, s, vt = np.linalg.svd(h)
#     v = vt.T

#     d = np.linalg.det(v @ u.T)
#     e = np.array([[1, 0, 0], [0, 1, 0], [0, 0, d]])

#     r = v @ e @ u.T
import matplotlib.colors as colors

def kabsch_rmsd(P, Q):
    """Calculate the RMSD between two sets of points using the Kabsch algorithm.

    Args:
        P (np.ndarray): Nx3 array of points
        Q (np.ndarray): Nx3 array of points

    Returns:
        float: RMSD between P and Q
    """
    P_save = P
    Q_save = Q
    P = P[:3]
    Q = Q[:3]
    
    # Center the data
    P -= np.mean(P, axis=0)
    Q -= np.mean(Q, axis=0)

    # Calculate the covariance matrix
    C = np.dot(np.transpose(P), Q)

    # Singular Value Decomposition
    V, S, W = np.linalg.svd(C)

    # Calculate the rotation matrix
    d = np.linalg.det(np.dot(V, W))
    R = np.dot(V, np.dot(np.array([[1,0,0], [0,1,0], [0,0,d]]), W))

    # Calculate the transformed coordinates
    P_transformed = np.dot(P_save, R)

    # Calculate the RMSD
    rmsd = np.sqrt(np.sum((Q_save - P_transformed)**2) / P_transformed.shape[0])

    return P_transformed, rmsd


# This one starts with two cross products to get a vector perpendicular to
# b2 and b1 and another perpendicular to b2 and b3. The angle between those vectors
# is the dihedral angle.
def dihedral3(p):
    # print(p)
    """formula from Wikipedia article on "Dihedral angle"; formula was removed
    from the most recent version of article (no idea why, the article is a
    mess at the moment) but the formula can be found in at this permalink to
    an old version of the article:
    https://en.wikipedia.org/w/index.php?title=Dihedral_angle&oldid=689165217#Angle_between_three_vectors
    uses 1 sqrt, 3 cross products"""
    p0 = p[0]
    p1 = p[1]
    p2 = p[2]
    p3 = p[3]

    b0 = -1.0*(p1 - p0)
    b1 = p2 - p1
    b2 = p3 - p2

    b0xb1 = np.cross(b0, b1)
    b1xb2 = np.cross(b2, b1)

    b0xb1_x_b1xb2 = np.cross(b0xb1, b1xb2)

    y = np.dot(b0xb1_x_b1xb2, b1)*(1.0/np.linalg.norm(b1))
    x = np.dot(b0xb1, b1xb2)

    return np.degrees(np.arctan2(y, x))


def bisector(dcm):
    vector1 = dcm[1] - dcm[0] #OH
    vector2 = dcm[1] - dcm[2] #OH
    unit_vector1 = vector1 / np.linalg.norm(vector1)
    unit_vector2 = vector2 / np.linalg.norm(vector2)
    bisector1 = (unit_vector1 + unit_vector2) / np.linalg.norm(
        unit_vector1 + unit_vector2)
    return bisector1


def angle(u,v):
    # Calculate polar angle (theta) between u and v
    cos_theta = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
    theta = np.degrees(np.arccos(cos_theta))
    return theta

#
# fig = plt.figure(figsize=(12, 12))
# ax = fig.add_subplot(projection='3d')
    

for idx, fnkey in enumerate(data_.coloumb.index):

    k = fnkey.split("_")[0]
    kab = fnkey.split(".")[0]
    s = structure_key_pairs[k]
    dcm_path = "/home/boittier/homeb/water_cluster/pbe0dz_pc/{}/charmm/dcm.xyz".format(k)
    s.load_dcm(dcm_path)
    dcms = np.array(s.dcm_charges)
    
    # calculate electrostatic energy
    E = 0
    dists = []
    pairs = [(int(kab.split("_")[1]), int(kab.split("_")[2]))]
    for pair in pairs:
        p1,p2 = pair
        mask1 = s.res_mask[p1]
        mask2 = s.res_mask[p2]
        dcm1 = dcms[mask1]
        dcm2 = dcms[mask2]
        dcm1c = dcm1.copy()
        dcm2c = dcm2.copy()
        dcm1 = dcm1[:,:-1]
        dcm2 = dcm2[:,:-1]
        
        if idx == 0:
            dcm1_save = dcm1
            dcm2_save = dcm2
            dcm12_save = np.concatenate([dcm1,dcm2])
        else:
            dcm1_2 = np.concatenate([dcm1,dcm2])
            dcms_transformed, rmsd = kabsch_rmsd(dcm1_2,dcm12_save)
            # print(rmsd)
            dcm1 = dcms_transformed[:3,:]
            dcm2 = dcms_transformed[3:,:]
            
        for i, a in enumerate(dcm1):
            for j, b in enumerate(dcm2):
                dist = np.linalg.norm(a[:-1] - b[:-1])
                Ec = Ecoloumb(a[-1],b[-1],dist)
                E += Ec
                if i == 0 and (j == 1 or j == 2):
                    dists.append(dist)
                if j == 0 and (i == 1 or i == 2):
                    dists.append(dist)
                
    CM1 = np.average(dcm1, axis=0, weights=[1,15.99,1])
    CM2 = np.average(dcm2, axis=0, weights=[1,15.99,1])
    distance = np.linalg.norm(CM1 - CM2)

    bisector1 = bisector(dcm1)
    bisector2 = bisector(dcm2)
    
#     print(dcm1.shape, dcm1)
#     print(dcm2.shape, dcm2)
    
    dih = dihedral3(np.array([dcm1[1,:]+bisector1, 
                              dcm1[1,:], 
                              dcm2[1,:], 
                              dcm2[1,:]+bisector2]))
    
    theta = angle(bisector1,bisector2)
    t11 = np.array([dcm1[1,:]+bisector1, 
                              dcm1[1,:]])
    t12 = np.array([dcm1[1,:], dcm2[1,:]])
    t21 = np.array([dcm2[1,:], dcm2[1,:]+bisector2])
    # t22 = np.array([dcm1[1,:]+bisector1, 
    #                           dcm1[1,:], 
    #                           dcm2[1,:], 
    #                           dcm2[1,:]+bisector2])
    
    theta_1 = angle(bisector1,bisector2)
    
    a1 = angle( dcm1[1,:] + bisector1, dcm1[1,:] + dcm2[1,:])
    a2 = angle( dcm2[1,:] + bisector2, dcm1[1,:] + dcm2[1,:] )
    if 120 - a1 - a2 < 0:
        # print(a1, a2, theta)
        theta = theta * -1
    
    # append data
    min_hbond.append(min(dists))
    E_col_dcms.append(E)
    dist_dcms.append(distance)
    angles_dcms.append(theta)
    azi_dcms.append(phi)
    dih_dcms.append(dih)
    

    THETA =  np.deg2rad(dih)
    PHI = np.deg2rad(theta)
    R = min(dists)
    #R = np.cos(THETA) * np.cos(THETA)
    X = 10*np.abs(R * np.sin(PHI) * np.cos(THETA))
    Y = 10*np.abs(R * np.sin(PHI) * np.sin(THETA))
    Z = 10*np.abs(R * np.cos(PHI))
    # X = 10*(R * np.sin(PHI) * np.cos(THETA))
    # Y = 10*(R * np.sin(PHI) * np.sin(THETA))
    # Z = 10*(R * np.cos(PHI))
    # dcm1 = dcm1 - np.array([X,Y,Z])
    dcm2 = dcm2 - np.array([X,Y,Z])
    
    dcms_.append([dcm1,dcm2])
    # print(idx, dih, theta)
    
    if distance < False:
        fig = plt.figure(figsize=(12, 12))
        ax = fig.add_subplot(projection='3d')
        norm = colors.Normalize(-12, 2)
        norm = colors.Normalize(0, 180)
        CMAP = plt.get_cmap('Reds_r')
        # CMAP = plt.get_cmap('rainbow')
        E = CMAP(norm(dih))

        # ax.scatter(X,Y,Z,c=E)
        # ax.scatter(*bisector2)
        # ax.plot(dcm1[:,])

        ax.plot(*dcm1[:,:].T,"--",c="k")
        ax.plot(*dcm2[:,:].T,c=E)
        # ax.plot(dcm1[0,:-1]+bisector1, dcm1[0,:-1])


        B1 = np.array([dcm1[1]+bisector1, dcm1[1]])
        ax.plot(*B1.T, "-",c=E)

        B2 = np.array([dcm2[1]+bisector2, dcm2[1]])
        ax.plot(*B2.T, "-",c=E)

        dih_coords = [dcm1[1]+bisector1, 
                                      dcm1[1], 
                                      dcm2[1], 
                                      dcm2[1]+bisector2]

        DIHP = np.array(dih_coords)
        ax.plot(*DIHP.T, "--", c="k")
        plt.title("{} : $\phi$ {:.2f} $\\theta$ {:.2f} a1 {:.2f} a2 {:.2f} \n {:.2f}".format(idx, dih, theta,
                                                                   a1, a2, 120 - a1 - a2), fontsize=20)

import matplotlib.cm as cm
sm = cm.ScalarMappable(cmap=CMAP,norm=norm)

plt.colorbar(sm)
plt.savefig("angles_dcm.pdf")

data_.coloumb["Ecol_dcm"] = E_col_dcms
data_.coloumb["dist_CM"] = dist_dcms
data_.coloumb["polar_angle"] = angles_dcms
data_.coloumb["azimuthal_angle"] = azi_dcms
data_.coloumb["min_hbond"] = min_hbond
data_.coloumb["dcm_coords"] = dcms_
```

```python
polar_angles = np.deg2rad(PC_dimer_anio['polar_angle'].values)
azimuthal_angles = np.deg2rad(PC_dimer_anio['dih_dcms'].values)

# Convert the angles to points on a unit sphere
x = np.sin(polar_angles) * np.cos(azimuthal_angles)
y = np.sin(polar_angles) * np.sin(azimuthal_angles)
z = np.cos(polar_angles)
points = np.column_stack((x, y, z))

# Calculate the centroid of the points
centroid = points.mean(axis=0)

# Calculate the angular deviation of each point from the centroid
cos_delta = np.dot(points, centroid)
delta = np.arccos(cos_delta)

# Calculate the length of the mean resultant vector
R = np.sqrt(np.sum(np.square(np.column_stack((x, y, z)))) / len(points))

# Calculate the normalization factor
k = -2 * np.log(1.000000000000001 - R)

# Normalize the angular deviation values
delta_norm = delta / k
# delta_norm = delta
# Save the normalized values to a new column in the dataframe
# plt.hist( np.rad2deg(delta_norm))
delta_norm_deg = np.rad2deg(delta_norm)
plt.hist(delta_norm_deg)
```
